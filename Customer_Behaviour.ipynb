```python
df.apply(lambda x:sum(x.isnull()),axis=0)
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
print(df.isnull().sum())
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
print(df.columns)
```

```python
labels=['Young','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
print(df[['age','age_group']].head(20))
```

```python
frequency_mapping={'Fortnightly':14, 'Weekly':7, 'Monthly':30, 'Quarterly':90, 'Bi-weekly':14, 'Annually':365, 'Every 3 months':90}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
print(df[['purchase_frequency_days','frequency_of_purchases']].head(20))
```
